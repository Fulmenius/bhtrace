```python
import pickle
import torch
import os
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
```

```python
# Define the directory to search in
directory = os.getcwd()
file_name = 'Sim_15.pickle'

with open(os.path.join(directory, file_name), 'rb') as file:
    # Do something with the file contents
    data = pickle.load(file)
```

```python
data['ResRN'].ys.shape
```

```python
k = 300 # 1 tp

# 151: [1 r, 2 th], 181: [1 r, 1 th] 

r_k = data['ResRN'].ys[k, :, 1]
mask1 = torch.greater(r_k, 2)
mask2 = torch.less(r_k, 10)
mask = torch.logical_and(mask1, mask2)

r_k = r_k[mask]
th_k = data['ResRN'].ys[k, :, 2][mask]
ph_k = data['ResRN'].ys[k, :, 3][mask]

# R-TH plot
fig, ax = plt.subplots(1,1, figsize=(6,6))

ax.plot(r_k, th_k/torch.pi, )

ax.yaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
# ax.yaxis.set_major_locator(tck.MultipleLocator(base=1.0))

# plt.style.use("ggplot")
plt.show()
```
